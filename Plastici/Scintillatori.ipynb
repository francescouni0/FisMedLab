```python
import matplotlib.pyplot as plt
import numpy as np

# Read data from a file

data_108 = np.loadtxt("F1--V_108_CFD_80_22cm_ch34.txt",delimiter=',')
data_110 = np.loadtxt("F1--V_110_CFD_80_22cm_ch34.txt",delimiter=',')
data_112 = np.loadtxt("F1--V_112_CFD_80_22cm_ch34.txt",delimiter=',')
data_114 = np.loadtxt("F1--V_114_CFD_80_22cm_ch34.txt",delimiter=',')
data_116 = np.loadtxt("F1--V_116_CFD_80_22cm_ch34.txt",delimiter=',')
data_118 = np.loadtxt("F1--V_118_CFD_80_22cm_ch34.txt",delimiter=',')
data_120 = np.loadtxt("F1--V_120_CFD_80_22cm_ch34.txt",delimiter=',')



# Extract times and counts
times_108 = data_108[:, 0]
counts_108 = data_108[:, 1]
times_110 = data_110[:, 0]
counts_110 = data_110[:, 1]
times_112 = data_112[:, 0]
counts_112 = data_112[:, 1]
times_114 = data_114[:, 0]
counts_114 = data_114[:, 1]
times_116 = data_116[:, 0]
counts_116 = data_116[:, 1]
times_118 = data_118[:, 0]
counts_118 = data_118[:, 1]
times_120 = data_120[:, 0]
counts_120 = data_120[:, 1]

# Plot histogram
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

plt.subplot(3, 3, 1)
plt.hist(times_108, weights=counts_108, bins=100)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("108")

plt.subplot(3, 3, 2)
plt.hist(times_110, weights=counts_110, bins=100)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("110")

plt.subplot(3, 3, 3)
plt.hist(times_112, weights=counts_112, bins=100)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("112")

plt.subplot(3, 3, 4)
plt.hist(times_114, weights=counts_114, bins=100)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("114")

plt.subplot(3, 3, 5)
plt.hist(times_116, weights=counts_116, bins=100)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("116")

plt.subplot(3, 3, 6)
plt.hist(times_118, weights=counts_118, bins=100)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("118")

plt.subplot(3, 3, 8)
plt.hist(times_120, weights=counts_120, bins=100)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("120")

# Adjust spacing
plt.tight_layout()

# Show plot
plt.show()
```

```python
from scipy.optimize import curve_fit
from scipy.stats import chisquare



# Define gaussian function
def gaussian(x, a, b, c):
    return a * np.exp(-((x - b) / c) ** 2)

V=[108,110,112,114,116,118,120]
counts=np.stack([counts_108,counts_110,counts_112,counts_114,counts_116,counts_118,counts_120],axis=0)
times=np.stack([times_108,times_110,times_112,times_114,times_116,times_118,times_120],axis=0)

print(np.shape(counts))

p0_list=[(95,-4.17995274e-11,0.3318e-9),(1.07605056e+02, -2.04905536e-11 , 3.03266025e-10),(9.76907093e+01, -9.02490361e-12,  2.91871235e-10),(9.76792576e+01, -7.66294285e-12,  2.88231386e-10),(1.97982742e+02 ,9.36570895e-12, 2.84337087e-10),(1.00941906e+02, 2.03798281e-12, 2.89832761e-10),(9.54787027e+01, 4.51361278e-12 ,2.96036116e-10)]

s=[]
ds=[]
T=[]
dT=[]

for i in range(7):
    
    
    sigma1=[np.var(counts[i,:])]*200
    # Fit gaussian function with curve_fit
    # Use bounds to fix b to 0 and let a and c vary
    popt, pcov = curve_fit(gaussian, times[i,:], counts[i,:],p0=p0_list[i],sigma=sigma1,absolute_sigma=False)
    
    # Print fitted parameters and covariance matrix
    print("Fitted parameters:", popt)
    print("Covariance matrix:", pcov)
    s=np.append(s,popt[2])
    T=np.append(T,popt[0])
    ds=np.append(ds,np.sqrt(pcov[2,2]))
    dT=np.append(dT,np.sqrt(pcov[0,0]))
    # Plot histogram and fitted function
    plt.hist(times[i,:], weights=counts[i,:], bins=200,)
    plt.xlabel("Time")
    plt.ylabel("Count")
    plt.title(f"Histogram of times{V[i]} and counts{V[i]}")
    
    x = np.linspace(min(times[i,:]), max(times[i,:]), 1000)
    y = gaussian(x, *popt)
    plt.plot(x, y, 'r-', label="Fitted curve")
    '''
    x_test = np.linspace(min(times[i,:]), max(times[i,:]), 1000)
    y_test = gaussian(x,1.75000000e+02,-0,2.15270818e-10 )
    plt.plot(x_test, y_test, 'y-', label="Fitted curve")
    plt.legend()
    plt.show()
    '''
    
    
    # Calculate expected frequencies from fitted function
    expected = gaussian(times[i,:], *popt)
    # Scale expected frequencies by ratio of sums
    ratio = sum(counts[i,:]) / sum(expected)
    expected = expected * ratio
    
    
    # Calculate chisquare and p-value
    chisq, p = chisquare(counts[i,:], expected)
    
    
    
    # Print results
    print("Chisquare:", chisq)
    print("P-value:", p)
    
    # Calculate chisquare reduced
    
    dof= len(times[i,:]) - 3
    print(dof)
    chisq_red = chisq / dof
    print("Chisquare reduced:", chisq_red)
    plt.show()
```
