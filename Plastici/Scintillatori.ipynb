# CTR al variare della OV, dati a 22 cm per CDF 80%

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.stats import chisquare
```

```python
# Read data from a file
data_108 = np.loadtxt("F1--V_108_CFD_80_22cm_ch34.txt",delimiter=',')
data_110 = np.loadtxt("F1--V_110_CFD_80_22cm_ch34.txt",delimiter=',')
data_112 = np.loadtxt("F1--V_112_CFD_80_22cm_ch34.txt",delimiter=',')
data_114 = np.loadtxt("F1--V_114_CFD_80_22cm_ch34.txt",delimiter=',')
data_116 = np.loadtxt("F1--V_116_CFD_80_22cm_ch34.txt",delimiter=',')
data_118 = np.loadtxt("F1--V_118_CFD_80_22cm_ch34.txt",delimiter=',')
data_120 = np.loadtxt("F1--V_120_CFD_80_22cm_ch34.txt",delimiter=',')

# Extract times and counts
times_108 = data_108[:, 0]
counts_108 = data_108[:, 1]
times_110 = data_110[:, 0]
counts_110 = data_110[:, 1]
times_112 = data_112[:, 0]
counts_112 = data_112[:, 1]
times_114 = data_114[:, 0]
counts_114 = data_114[:, 1]
times_116 = data_116[:, 0]
counts_116 = data_116[:, 1]
times_118 = data_118[:, 0]
counts_118 = data_118[:, 1]
times_120 = data_120[:, 0]
counts_120 = data_120[:, 1]

# Plot histogram
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

plt.subplot(3, 3, 1)
plt.hist(times_108, weights=counts_108, bins=200)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("SiPM @ 108V")

plt.subplot(3, 3, 2)
plt.hist(times_110, weights=counts_110, bins=200)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("SiPM @ 110V")

plt.subplot(3, 3, 3)
plt.hist(times_112, weights=counts_112, bins=200)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("SiPM @ 112V")

plt.subplot(3, 3, 4)
plt.hist(times_114, weights=counts_114, bins=200)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("SiPM @ 114V")

plt.subplot(3, 3, 5)
plt.hist(times_116, weights=counts_116, bins=200)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("SiPM @ 116V")

plt.subplot(3, 3, 6)
plt.hist(times_118, weights=counts_118, bins=200)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("SiPM @ 118V")

plt.subplot(3, 3, 8)
plt.hist(times_120, weights=counts_120, bins=200)
plt.xlabel("Time [s]")
plt.ylabel("Count")
plt.title("SiPM @ 120V")

# Adjust spacing
plt.tight_layout()

# Show plot
plt.show()
```

### Fit gaussiano delle distribuzioni

```python
# Define gaussian function
def gaussian(x, a, b, c):
    return a * np.exp(-0.5*((x - b) / c) ** 2)

def sigma_start(c, t):
    over_half = np.nonzero((c > c.max()/2))[0]
    bc = (t[1:]+t[:-1])/2
    FWHM = bc[over_half[-1]] - bc[over_half[0]]
    sigma = FWHM/3.35
    return sigma

V=[108,110,112,114,116,118,120]
# Concatenazione verticale: ogni riga  un OV, le colonne sono i conteggi per ogni bin / il tempo corrispondente al bin
counts=np.stack([counts_108,counts_110,counts_112,counts_114,counts_116,counts_118,counts_120],axis=0)
times=np.stack([times_108,times_110,times_112,times_114,times_116,times_118,times_120],axis=0)

# SONO OTTIMIZZABILI
# p0_list=[(95,-4.17995274e-11,0.3318e-9),(1.07605056e+02, -2.04905536e-11 , 3.03266025e-10),(9.76907093e+01, -9.02490361e-12,  2.91871235e-10),(9.76792576e+01, -7.66294285e-12,  2.88231386e-10),(9.54787027e+01, 4.51361278e-12 ,2.96036116e-10),(1.00941906e+02, 2.03798281e-12, 2.89832761e-10),(9.54787027e+01, 4.51361278e-12 ,2.96036116e-10)]
p0_list=[(counts[0,:].max(), np.average(times[0,:], weights=counts[0,:]), sigma_start(counts[0,:],times[0,:])),
         (counts[1,:].max(), np.average(times[1,:], weights=counts[1,:]), sigma_start(counts[1,:],times[1,:])),
         (counts[2,:].max(), np.average(times[2,:], weights=counts[2,:]), sigma_start(counts[2,:],times[2,:])),
         (counts[3,:].max(), np.average(times[3,:], weights=counts[3,:]), sigma_start(counts[3,:],times[3,:])),
         (counts[4,:].max(), np.average(times[4,:], weights=counts[4,:]), sigma_start(counts[4,:],times[4,:])),
         (counts[5,:].max(), np.average(times[5,:], weights=counts[5,:]), sigma_start(counts[5,:],times[5,:])),
         (counts[6,:].max(), np.average(times[6,:], weights=counts[6,:]), sigma_start(counts[6,:],times[6,:]))]

s=[]
ds=[]
T=[]
dT=[]

for i in range(7):
    sigma1=[np.var(counts[i,:])]*200
    # Fit gaussian function with curve_fit
    # Use bounds to fix b to 0 and let a and c vary
    popt, pcov = curve_fit(gaussian, times[i,:], counts[i,:],p0=p0_list[i],sigma=sigma1,absolute_sigma=False)
    
    # Print fitted parameters and covariance matrix
    print("Fitted parameters:\n", popt)
    print("Covariance matrix:\n", pcov)
    s=np.append(s,popt[2])
    T=np.append(T,popt[0])
    ds=np.append(ds,np.sqrt(pcov[2,2]))
    dT=np.append(dT,np.sqrt(pcov[0,0]))

    # Plot histogram and fitted function
    plt.hist(times[i,:], weights=counts[i,:], bins=200,)
    plt.xlabel("Time [s]")
    plt.ylabel("Count")
    plt.title(f"SiPM @ {V[i]}V")
    
    x = np.linspace(min(times[i,:]), max(times[i,:]), 1000)
    y = gaussian(x, *popt)
    plt.plot(x, y, 'r-', label="Fitted curve")

    '''
    x_test = np.linspace(min(times[i,:]), max(times[i,:]), 1000)
    y_test = gaussian(x,1.75000000e+02,-0,2.15270818e-10 )
    plt.plot(x_test, y_test, 'y-', label="Fitted curve")
    plt.legend()
    plt.show()
    '''

    # Calculate expected frequencies from fitted function
    expected = gaussian(times[i,:], *popt)
    # Scale expected frequencies by ratio of sums
    ratio = sum(counts[i,:]) / sum(expected)
    expected = expected * ratio

    # Calculate chisquare and p-value
    chisq, p = chisquare(counts[i,:], expected)

    # Print results
    print("Chi Square:\n", chisq)
    print("P-value:\n", p)
    
    # Calculate chisquare reduced
    dof= len(times[i,:]) - 3
    print("DoF:\n", dof)
    chisq_red = chisq / dof
    print("Chi Square Reduced:\n", chisq_red)
    plt.show()
```

```python
# Define gaussian function
def gaussian(x, a, b, c):
    return a * np.exp(-((x - b) / c) ** 2)

def sigma_start(c, t):
    over_half = np.nonzero((c > c.max()/2))[0]
    bc = (t[1:]+t[:-1])/2
    FWHM = bc[over_half[-1]] - bc[over_half[0]]
    sigma = FWHM/3.35
    return sigma

D=[4,8,12,16,20,22,24,28,32,36,40]
counts=np.stack([counts_4,counts_8,counts_12,counts_16,counts_20,counts_22,counts_24,counts_28,counts_32,counts_36,counts_40],axis=0)
times=np.stack([times_4,times_8,times_12,times_16,times_20,times_22,times_24,times_28,times_32,times_36,times_40],axis=0)

print(np.shape(counts))
#SONO OTTIMIZZABILI
#p0_list=[(100,-2.5e-9,0.3318e-9),(1.07605056e+02, -1.8e-9 , 3.03266025e-10),(9.76907093e+01, -1.4e-9,  2.91871235e-10),(9.76792576e+01, -0.7e-9,  2.88231386e-10),(9.54787027e+01, -0.25e-9 ,2.96036116e-10),(200,0, 2.89832761e-10),(9.54787027e+01,0.25e-9 ,2.96036116e-10),(9.54787027e+01,0.25e-9 ,2.96036116e-10),(9.54787027e+01, 0.3e-9 ,2.96036116e-10),(9.54787027e+01, 0.2e-9 ,2.96036116e-10),(120, 0.25e-9 ,2.96036116e-10)]
p0_list=[(counts[0,:].max(), np.average(times[0,:], weights=counts[0,:]), sigma_start(counts[0,:],times[0,:])),
         (counts[1,:].max(), np.average(times[1,:], weights=counts[1,:]), sigma_start(counts[1,:],times[1,:])),
         (counts[2,:].max(), np.average(times[2,:], weights=counts[2,:]), sigma_start(counts[2,:],times[2,:])),
         (counts[3,:].max(), np.average(times[3,:], weights=counts[3,:]), sigma_start(counts[3,:],times[3,:])),
         (counts[4,:].max(), np.average(times[4,:], weights=counts[4,:]), sigma_start(counts[4,:],times[4,:])),
         (counts[5,:].max(), np.average(times[5,:], weights=counts[5,:]), sigma_start(counts[5,:],times[5,:])),
         (counts[6,:].max(), np.average(times[6,:], weights=counts[6,:]), sigma_start(counts[6,:],times[6,:])),
         (counts[7,:].max(), np.average(times[7,:], weights=counts[7,:]), sigma_start(counts[7,:],times[7,:])),
         (counts[8,:].max(), np.average(times[8,:], weights=counts[8,:]), sigma_start(counts[8,:],times[8,:])),
         (counts[9,:].max(), np.average(times[9,:], weights=counts[9,:]), sigma_start(counts[9,:],times[9,:])),
         (counts[10,:].max(), np.average(times[10,:], weights=counts[10,:]), sigma_start(counts[10,:],times[10,:])),]

s=[]
ds=[]
T=[]
dT=[]

for i in range(11):
    sigma1=[np.var(counts[i,:])]*200
    # Fit gaussian function with curve_fit
    # Use bounds to fix b to 0 and let a and c vary
    popt, pcov = curve_fit(gaussian, times[i,:], counts[i,:],p0=p0_list[i],sigma=sigma1,absolute_sigma=False)
    
    # Print fitted parameters and covariance matrix
    print("Fitted parameters:\n", popt)
    print("Covariance matrix:\n", pcov)
    s=np.append(s,popt[2])
    T=np.append(T,popt[1])
    ds=np.append(ds,np.sqrt(pcov[2,2]))
    dT=np.append(dT,np.sqrt(pcov[1,1]))

    # Plot histogram and fitted function
    plt.hist(times[i,:], weights=counts[i,:], bins=200,color='c')
    plt.xlabel("Time [s]")
    plt.ylabel("Count")
    plt.title(f"Source position @ {D[i]} cm")
    
    x = np.linspace(min(times[i,:]), max(times[i,:]), 1000)
    y = gaussian(x, *popt)
    plt.plot(x, y,'orange', label="Fitted curve")

    '''
    x_test = np.linspace(min(times[i,:]), max(times[i,:]), 1000)
    y_test = gaussian(x,1.75000000e+02,-0,2.15270818e-10 )
    plt.plot(x_test, y_test, 'y-', label="Fitted curve")
    plt.legend()
    plt.show()
    '''

    # Calculate expected frequencies from fitted function
    expected = gaussian(times[i,:], *popt)
    # Scale expected frequencies by ratio of sums
    ratio = sum(counts[i,:]) / sum(expected)
    expected = expected * ratio

    # Calculate chisquare and p-value
    chisq, p = chisquare(counts[i,:], expected)

    # Print results
    print("Chi Square:\n", chisq)
    print("P-value:\n", p)
    
    # Calculate chisquare reduced
    dof= len(times[i,:]) - 3
    print("DoF:\n",dof)
    chisq_red = chisq / dof
    print("Chi Square Reduced:\n", chisq_red)
    plt.show()
```
