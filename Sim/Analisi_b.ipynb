```python
from scipy.optimize import curve_fit    

def gaussian(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

tm200=(m200_1.iloc[:,2]-m200_2.iloc[:,2])
tm200=tm200


n,bins,_=plt.hist(tm200,bins='auto', alpha=0.5, label='Histogram')

bin_centers = (bins[:-1] + bins[1:]) / 2


# Fit the histogram using curve_fit
fit_params, _ = curve_fit(gaussian, bin_centers, n,p0=[np.max(n),np.mean(tm200),np.std(tm200)])

# Generate x values for plotting the fitted distribution
x = np.linspace(np.min(bin_centers), np.max(bin_centers), 100000)


# Plot the fitted distribution
plt.plot(x, gaussian(x, *fit_params), 'r-', label='Fitted Distribution')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
'''
plt.plot(x,gaussian(x,20000,0,0.15), 'g-', label='prova')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
'''

print(fit_params)
mean=np.append(mean,fit_params[1])
mean_1=np.append(mean_1,np.mean(tm200))
std=np.append(std,fit_params[2])  

std_1=np.append(std_1,np.std(tm200))  
```
