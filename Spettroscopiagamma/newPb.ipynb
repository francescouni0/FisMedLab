```python
import numpy as np 
import matplotlib.pyplot as plt
from pathlib import Path
from scipy.signal import find_peaks
from scipy.optimize import curve_fit

Path.cwd()
```

```python
Cs137_0 = np.loadtxt('Cs137_0mm_2.Spe')
Cs137_4= np.loadtxt('Cs137_4mm.Spe')
Cs137_6= np.loadtxt('Cs137_6mm.Spe')
Cs137_9= np.loadtxt('Cs137_9mm.Spe')
Cs137_200= np.loadtxt('Cs137_20cm.Spe')
Cs137_100= np.loadtxt('Cs137_1cm.Spe')

back=np.loadtxt('fondo.Spe')
back=back/ (np.max(back))


Cs137_0=Cs137_0-back
Cs137_4=Cs137_4-back
Cs137_6=Cs137_6-back
Cs137_9=Cs137_9-back
Cs137_200=Cs137_200-back
Cs137_100=Cs137_100-back

A_net=[]
```

```python

def calculate_chi_square(y_observed, y_expected, uncertainties):
    residuals = y_observed - y_expected
    chi_square = np.sum((residuals / uncertainties) ** 2)
    return chi_square
```

```python
x=np.linspace(1,2048,2048)


energy=x

intensity=Cs137_0



Ca_index = 700
Cb_index = 900

Ca_intensity = intensity[Ca_index]
Cb_intensity = intensity[Cb_index]

continuum_slope = (Cb_intensity - Ca_intensity) / (Cb_index - Ca_index)

continuum_approximation = Ca_intensity + continuum_slope * (energy - energy[Ca_index])
continuum_approximation[:Ca_index] = 0
continuum_approximation[Cb_index:] = 0

subtracted_spectrum = intensity - continuum_approximation
subtracted_spectrum[:Ca_index] = 0
subtracted_spectrum[Cb_index:] = 0


A_net.append(np.sum(subtracted_spectrum[Ca_index:Cb_index]))
print(A_net)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(energy, intensity, label='Original Spectrum')
plt.xlabel('Energy')
plt.ylabel('Intensity')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(energy[Ca_index:Cb_index], continuum_approximation[Ca_index:Cb_index], label='Continuum Approximation')
plt.plot(energy, intensity, label='Original Spectrum')
plt.xlabel('Energy')
plt.ylabel('Intensity')
plt.legend()

```

```python
A_net=np.array(A_net)
def exp_fit(x, a, b):
    return a * np.exp(-b * x) 

x=np.array([0,0.4,0.6,0.9,1])
dy=[A_net[0]*0.08,A_net[1]*0.08,A_net[2]*0.08,A_net[3]*0.08,A_net[4]*0.08]

popt, pcov = curve_fit(exp_fit, x, A_net, sigma=dy, absolute_sigma=False,p0=[3500,1])

# Retrieve the optimized parameters
a_fit, b_fit= popt

# Generate points for the fitted exponential curve
x_fit = np.linspace(0, 1, 100)
y_fit = exp_fit(x_fit, a_fit, b_fit)

# Plot the original data and the fitted curve
plt.figure(figsize=(10, 10))
plt.errorbar(x, A_net,yerr=dy, label='Data',fmt='x',color='b')
plt.plot(x_fit, y_fit, 'r', label='Fitted curve',color='m',)
plt.xlabel('Thickness (cm)',fontsize=20)
plt.ylabel('Net Area',fontsize=20)
plt.legend()

plt.show()

print(popt)
print(np.sqrt(np.diag(pcov)))

print((popt[1]*0.001))
fity=exp_fit(x,popt[0],popt[1])

print(calculate_chi_square(A_net,fity,dy))
```
